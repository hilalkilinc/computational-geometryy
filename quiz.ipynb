```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python

def my_f(plane):
    point  = np.array([0, 0, 0])
    normal = plane
   
    point1  = np.array([-10, 0, 0])
    normal1 = plane
   
    d1 = -point.dot(normal)
    d2 = -point1.dot(normal1)
   
    xx, yy = np.meshgrid(range(10), range(10))
   
    z1 = (-normal[0] * xx - normal[1] * yy - d1) * 1. /normal[2]
    z2 = (-normal1[0] * xx - normal1[1] * yy - d2) * 1. /normal1[2]
   
    uzakl覺k = abs(point[0]-point1[0]) / int((normal[0]**2 + normal[1]**2 + normal[2]**2)**.5)
   
    print("uzakl覺k: ", uzakl覺k)
   
    plt3d = plt.figure().gca(projection='3d')
    plt3d.plot_surface(xx, yy, z1, color='blue')
    plt3d.plot_surface(xx, yy, z2, color='blue')
    plt3d.plot([point[1],point1[2]], [point[1],point1[1]], [point[0],point1[0]])
    plt.show()
 
```

```python
plane=np.array([1,1,1])
my_f(plane)
```
